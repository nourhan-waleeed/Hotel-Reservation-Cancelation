# 1.Problem Statement:
   **Is to analyze hotel booking data to understand booking patterns,and factors affecting cancellations/non-cancellations, aiming to optimize hotel operations and enhance customer experience.**

# 2.Data Description
#### 1)Booking_ID: unique identfier for each booking 
#### 2)Number of Adults: The count of adults included in the booking.
#### 3)Number of Children: The count of children included in the booking.
#### 4)Number of Weekend Nights: The number of nights that fall on the weekend.
#### 5)Number of Week Nights: The number of nights that fall on weekdays.
#### 6)Type of Meal: The meal plan (Meal Plan 1, Meal Plan 2,Meal Plan 3, Meal Plan4, or Not Selected).
#### 7)Car Parking Space: Indicates whether a car parking space was requested or not( binary).
#### 8)Room Type: The type of room (Room_Type 1, Room_Type 4).
#### 9)Lead Time: The number of days between the date of reservation and the date of the booking.
#### 10)Market Segment Type: Did the cutomer book online or offline 
#### 11)Repeated:indicate whether its a new cutomer (0) or made a reservation before(0).
#### 12)P-C (Previous Cancellations): Indicates the number cancellations.
#### 13)P-not-C: indicates the number of not canceled reservations.
#### 14)Average Price: avg price of the reservation by each customer.
#### 15)Special Requests: The number of special requests made.
#### 16)Date of Reservation: The date when the reservation was made.
#### 17)Booking Status: The status of the booking (e.g., Not_Canceled, Canceled).

```python
df =booking.copy()
```

```python
df
```

```python
df['type of meal'] = df['type of meal'].str.replace('Meal Plan ', '')
df['room type'] = df['room type'].str.replace('Room_Type ', '')
```

```python

encoder = OneHotEncoder(handle_unknown='ignore')

enc_data = pd.DataFrame(encoder.fit_transform(df[['type of meal']]).toarray(), columns = 'type of meal'+'_'+encoder.categories_[0])
df = pd.concat([df.reset_index(drop=True), enc_data.reset_index(drop=True)], axis=1)
df.drop(columns='type of meal', inplace=True)
```

```python
encoder2 = OneHotEncoder(handle_unknown='ignore')

enc_data = pd.DataFrame(encoder2.fit_transform(df[['room type']]).toarray(), columns = 'room type'+'_'+encoder2.categories_[0])
df = pd.concat([df.reset_index(drop=True), enc_data.reset_index(drop=True)], axis=1)
df.drop(columns='room type', inplace=True)
```

```python
encoder3 = OneHotEncoder(handle_unknown='ignore')

enc_data = pd.DataFrame(encoder3.fit_transform(df[['market segment type']]).toarray(), columns = 'market segment type'+'_'+encoder3.categories_[0])
df = pd.concat([df.reset_index(drop=True), enc_data.reset_index(drop=True)], axis=1)
df.drop(columns='market segment type', inplace=True)
```
